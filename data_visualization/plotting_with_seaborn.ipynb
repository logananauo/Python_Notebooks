### **Visualization with Seaborn**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid", palette="rainbow", context="notebook")
```

```python
sns.get_dataset_names()
```

```python
### seaborn planets dataset
penguins = sns.load_dataset('penguins').dropna()
penguins.head()
```

```python
### gapminder dataset; life expectancy 
gap_url = 'https://raw.githubusercontent.com/chendaniely/pandas_for_everyone/master/data/gapminder.tsv'
gap = pd.read_csv(gap_url, sep='\t')
gap.head()
```

#### **Scatterplots:**

```python
### figure-level
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='scatter', alpha=0.35, s=200, edgecolor='none')
plt.show()
```

```python
### incorporating categorical variable
sns.relplot(data=penguins,
            x='bill_length_mm',
            y='bill_depth_mm',
            kind='scatter',
            hue='species',
            col='island',
            row='sex',
            alpha=0.35,
            s=100,
            edgecolor='none',
            height=4)
plt.show()
```

```python
### axes-level
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', alpha=0.7, s=100, hue='island')
plt.title('Bill Depth vs. Bill Length')
plt.xlabel('Bill Depth')
plt.ylabel('Bill Length')
plt.show()
```

```python
### multiple scatter plot
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', alpha=0.7, s=50, hue='sex', ax=ax[0])
ax[0].set_title('Bill Depth vs. Bill Length')
ax[0].set_xlabel('Bill Length')
ax[0].set_ylabel('Bill Depth')

sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', alpha=0.7, s=50, hue='sex', ax=ax[1])
ax[1].set_title('Flipper Length vs. Bill Length')
ax[1].set_xlabel('Bill Length')
ax[1].set_ylabel('Flipper Length')

sns.scatterplot(data=penguins, x='bill_length_mm', y='body_mass_g', alpha=0.7, s=50, hue='sex', ax=ax[2])
ax[2].set_title('Body Mass vs. Bill Length')
ax[2].set_xlabel('Bill Length')
ax[2].set_ylabel('Flipper Length')

plt.tight_layout()
plt.show()
```

#### **Kernel Density Estimate (KDE):**

```python
sns.displot(data=penguins, x='bill_depth_mm', kind='kde', height=4)
plt.show()
```

```python
### removing the sample size effect
sns.displot(data=penguins, x='bill_depth_mm', hue='species', kind='kde', common_norm=False, height=4)
plt.show()
```

#### **Grouped Bar Charts:**

#### **Heatmaps:**

```python
ct = pd.crosstab(penguins['species'], penguins['island'])
sns.heatmap(ct, fmt='d', annot=True)
plt.show()
```

#### **Boxplots:**

```python
### figure-level
sns.catplot(data=penguins, x='species', y='flipper_length_mm', kind='box', showmeans=True, height=5)
plt.show()
```

#### **Violin Plots:**

```python
sns.catplot(data=penguins, x='island', y='flipper_length_mm', kind='violin')
plt.show()
```

#### **Point Plots:**

```python
### figure-level
sns.catplot(data=gap, x='year', y='lifeExp', kind='point', linestyle='none', height=5)
plt.xticks(rotation=50)
plt.show()
```

```python
### stylize the plot
sns.catplot(
    data=gap,
    x='year',
    y='lifeExp',
    kind='point',
    linestyle='none',
    capsize=0.1,
    err_kws={'linewidth':1},
    height=4,
    aspect=3
)
plt.show()
```

```python
### compare 3 variables; 1 continuous and 2 categorical
sns.catplot(
    data=gap,
    x='year',
    y='lifeExp',
    hue='continent',
    kind='point',
    linestyle='--',
    capsize=0.1,
    errorbar=('ci', 68),
    err_kws={'linewidth':1},
    height=4,
    aspect=3
)
plt.show()
```

#### **Pairplot:**

```python
sns.pairplot(data=penguins, hue='species')
plt.show()
```

```python
### removing sample size effect on diagonal kde plots
sns.pairplot(data=penguins, hue='island', diag_kws={'common_norm':False})
plt.show()
```
